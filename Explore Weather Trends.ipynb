## Extract Data

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

## Read the data

```python
data = pd.read_csv('results.csv')
```

```python
data.head()
```

## Create the line chart:

Compare local and global moving average of average temperature trend.

```python
local_rolling = data.city_temp.rolling(window = 10).mean()
global_rolling = data.global_temp.rolling(window = 10).mean()
```

basic

```python
plt.plot(data.year, local_rolling, color='blue', label = 'local temperature')
plt.plot(data.year, global_rolling, color='red', label = 'global temperature')
plt.xlabel('year')
plt.ylabel('average temperature')
plt.title('Local and global moving average of average temperature trend')
plt.legend()
plt.show()
```

with a reference line

```python
plt.plot(data.year, local_rolling, color='blue', label = 'San Francisco temperature')
plt.plot(data.year, global_rolling, color='red', label = 'global temperature')
plt.xlabel('year')
plt.ylabel('average temperature (â„ƒ)')
plt.title('local and global moving average of average temperature trend')

plt.plot([1860, 2020], [data.city_temp.mean(), data.city_temp.mean()], 'k-', lw=1,dashes=[2, 2])
plt.text(2008, 14.5, str(data.city_temp.mean().round(2)) + 'â„ƒ')
plt.text(2000, 14, 'St.Dev:' +  str(data.city_temp.std().round(2)))

plt.plot([1860, 2020], [data.global_temp.mean(), data.global_temp.mean()], 'k-', lw=1,dashes=[2, 2])
plt.text(2010, 8.5, str(data.global_temp.mean().round(2)) + 'â„ƒ')
plt.text(1997, 8, 'St.Dev:' + str(data.global_temp.std().round(2)))
         
plt.legend(bbox_to_anchor=(1.04,0), loc="lower left")
plt.show()
```
